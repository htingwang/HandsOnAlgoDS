# 694. Number of Distinct Islands
#### https://leetcode.com/problems/number-of-distinct-islands/

# Approach 1: BFS

```python
from collections import deque
class Solution(object):
    def numDistinctIslands(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        if not grid or not grid[0]: return 0
        m, n = len(grid), len(grid[0])
        island = set()
        for i in range(m):
            for j in range(n):
                cur_island = []
                if grid[i][j] == 1:
                    queue = deque()
                    queue.append((i, j))
                    grid[i][j] = 0
                    cur_island.append(0)
                    while queue:
                        cur_i, cur_j= queue.popleft()
                        
                        for idx, (ni, nj) in enumerate([(cur_i - 1, cur_j), (cur_i + 1, cur_j), (cur_i, cur_j - 1), (cur_i, cur_j + 1)]):
                            if 0 <= ni < m and 0 <= nj < n and grid[ni][nj] == 1:
                                queue.append((ni, nj))
                                cur_island.append(idx + 1)
                                grid[ni][nj] = 0
                        cur_island.append(0)
                if cur_island: island.add(tuple(cur_island))
        return len(island)
```

# Approach 2: DFS

```python
class Solution(object):
    def numDistinctIslands(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        if not grid or not grid[0]: return 0
        m, n = len(grid), len(grid[0])
        island = set()
        
        def dfs(i, j, di):
            if 0 <= i < m and 0 <= j < n and grid[i][j] == 1:
                grid[i][j] = 0
                shape.append(di)
                dfs(i + 1, j, 1)
                dfs(i - 1, j, 2)
                dfs(i, j + 1, 3)
                dfs(i, j - 1, 4)
                shape.append(0)
        
        for i in range(m):
            for j in range(n):
                shape = []
                dfs(i, j, 0)
                if shape: island.add(tuple(shape))
        return len(island)
```
