# 621. Task Scheduler
#### https://leetcode.com/problems/task-scheduler/

# Approach 1

```python
class Solution(object):
    def leastInterval(self, tasks, n):
        """
        :type tasks: List[str]
        :type n: int
        :rtype: int
        """
        mapping = [0] * 26
        for task in tasks:
            mapping[ord(task) - ord("A")] += 1
        mapping.sort(reverse = True)
        
        res = 0
        while mapping[0] > 0:
            i = 0
            while i <= n:
                if mapping[0] == 0:
                    break
                if i < len(mapping):
                    mapping[i] -= 1
                res += 1
                i += 1
            mapping.sort(reverse = True)
            
        return res
```
